```python
import pandas as pd
```

# 1. 대학 목록
- 전국 대학교 목록 조회
- 서울시 내 대학교로 제한

```python
# https://www.data.go.kr/data/15100330/fileData.do
university = pd.read_excel('/Users/donghyunkim/Downloads/학교개황(20220310기준).xls')
university.head()
```

```python
# 서울 내 대학교, 전문대학교로 한정
university_seoul = university[((university['학교구분'] == '대학') | (university['학교구분'] == '전문대학')) & (university['지역'] == '서울')]
university_seoul = university_seoul.replace('케이씨대학교', '강서대학교')
university_seoul.head()
```

# 2. 학생 목록
- 학교 및 캠퍼스별 학과 수, 학생 수 불러오기
- 재학생 수로 한정
- 학교 목록의 학교명, 캠퍼스명 통일 후 조인하여 파일 합치기

```python
# https://kess.kedi.re.kr/contents/dataset
student = pd.read_excel('/Users/donghyunkim/Downloads/2023년 고등 학교별 학과수 입학정원 지원 입학 학생 외국인학생 졸업 교직.xlsx')
student.head()
```

```python
# 본분교명 수정
student = student.replace({'본교(제1캠퍼스)' : '본교', '본교(제2캠퍼스)' : '제2캠퍼스', '본교(제3캠퍼스)': '제3캠퍼스', '명지대학교 인문캠퍼스' : '명지대학교', 
                           '한국폴리텍대학 서울정수캠퍼스' : '한국폴리텍 I 대학 서울정수캠퍼스', '한국폴리텍대학 서울강서캠퍼스' : '한국폴리텍 I 대학 서울강서캠퍼스'})[['학교명', '본분교', '재학생_전체_계', '재학생_전체_남', '재학생_전체_여']]
university_seoul = pd.merge(left = university_seoul, right = student, how = 'left', on = ['학교명', '본분교'])
```

# 3. 학교 보유면적
- 대학교 별 보유면적 불러오기
- 학교 목록의 학교명, 캠퍼스명 통일 후 조인하여 파일 합치기
- 데이터셋에 없는 대학교는 따로 탐색하여 데이터 추가


```python
# https://www.academyinfo.go.kr/uipnh/unt/unmcom/RdViewer.do
area = pd.read_excel('/Users/donghyunkim/Downloads/2022년_교육여건_교지+확보율.xls')
area
```

```python
# 본분교명 수정
area.rename(columns = {'캠퍼스명' : '본분교', '보유면적(㎡)(C)' : '보유면적'}, inplace = True)
area['본분교'] = area['본분교'].str.split('(', expand = True)[0]
area = area.replace({'인문캠퍼스' : '본교', '성신캠퍼스' : '제3캠퍼스', '성의캠퍼스' : '제2캠퍼스', '쌍문캠퍼스' : '본교', 
                     '종로캠퍼스' : '제2캠퍼스', '인문사회과학캠퍼스' : '본교', '운정그린캠퍼스' : '제2캠퍼스'})
area = area[['학교명', '본분교', '보유면적']]
```

```python
# 조인 및 사이버대학 제외, 기타 값 추가
university_seoul = pd.merge(left = university_seoul, right = area, how = 'left', on = ['학교명', '본분교'])
university_seoul.iloc[18, -1] = 55486 # 명지대학교
university_seoul = university_seoul[university_seoul['학제'] != '사이버대학(대학)']
university_seoul = university_seoul[university_seoul['학교명'] != '한국방송통신대학교']
university_seoul = university_seoul[university_seoul['학교명'] != '정석대학']
university_seoul.iloc[43, -1] = 143768 # 한국예술종합학교 https://academyinfo.go.kr/uipnh/unt/unmcom/RdViewer.do
university_seoul.iloc[-3, -1] = 25326 # 한국폴리텍 정수캠퍼스https://academyinfo.go.kr/uipnh/unt/unmcom/RdViewer.do
university_seoul.iloc[-2, -1] = 32403 # 한국폴리텍 강서캠퍼스 https://academyinfo.go.kr/uipnh/unt/unmcom/RdViewer.do
```

```python
address = university_seoul['주소'].str.split(' ', expand = True)
address.columns = ['시', '구', '로', '번', '동', '대', '기타']
address = address[['시', '구', '동']]
address['동'] = address['동'].str[1:-1]
```

```python
university_seoul = pd.concat([university_seoul, address], axis = 1)
university_seoul['지번주소'] = university_seoul['시'] + " " + university_seoul['구'] + " " + university_seoul['동']
university_seoul.head()
```
